```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips=pd.read_csv('/Users/anupprakash/Documents/ExcelRDataScience/DataSetFile/tips.csv')
tips.head(10)
```

```python
plt.figure(figsize=(18,9))
sns.boxplot(y='total_bill',data=tips)

plt.show()
```

```python
plt.figure(figsize=(15,8))


sns.boxplot(x='day',y='total_bill',data=tips)


plt.show()
```

```python
plt.figure(figsize=(15,8))


sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')


plt.show()
```

```python
plt.figure(figsize=(15,8))


sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',palette='husl')


plt.show()
```

```python
plt.figure(figsize=(15,8))


sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',palette='husl',order=['Fri','Thur','Sat','Sun'])


plt.show()
```

```python
plt.figure(figsize=(15,8))


sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',palette='husl',order=['Fri','Thur','Sat','Sun'])
sns.swarmplot()

plt.show()
```
