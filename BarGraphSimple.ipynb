```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
languages = ['Python','Java','R','Koltin','C','C++','Ruby','Perl','Scala','Go']
student1 = np.random.randint(9,90,10)
student2 = np.random.randint(4,40,10)
```

```python
print(student1)
print(student2)
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.bar(languages,student1,color='r',width=0.92,align='center',label='Languages Student ratio')
# align is use place the bars in centre or left or right

plt.title('Number of student in Language Class')
plt.xlabel('Languages')
plt.ylabel('Students')

#plt.axis([8])
plt.legend(loc=2)

plt.grid(color='k',linestyle=':',linewidth=4)
plt.show()
```
