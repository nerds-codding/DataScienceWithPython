```python
from matplotlib import style
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
cd=pd.read_csv('/Users/anupprakash/Desktop/Assignments/Assignment3/crime_data.csv')
cd.head()
```

```python
cd.rename(columns={ cd.columns[0]: "City" }, inplace = True)
```

```python
x = cd.City
y1 = cd.Murder
y2 = cd.Assault
y3 = cd.UrbanPop
y4 = cd.Rape
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.grid(color='k',linestyle=':')

plt.scatter(x,y1,color='m',label='Murder',marker='o')

plt.xlabel('City')
plt.ylabel('Crime')
plt.title('Crime Data')

plt.legend(loc=1)

plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.grid(color='k',linewidth=1)

plt.scatter(x,y2,color='b',label='Assault',marker='<')

plt.xlabel('City')
plt.ylabel('Assault')
plt.title('Crime Data')

plt.legend(loc=1)

plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.grid(color='k',linestyle=':')

plt.scatter(x,y3,color='m',label='UrbanPop',marker='o')

plt.xlabel('City')
plt.ylabel('UrbanPop')
plt.title('Crime Data')

plt.legend(loc=1)

plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.grid(color='k',linestyle=':')

plt.scatter(x,y4,color='b',label='Rape',marker='<')

plt.xlabel('City')
plt.ylabel('Rape')
plt.title('Crime Data')

plt.legend(loc=1)

plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.grid(color='k',linestyle=':')

plt.scatter(x,y1,color='b',label='Mruder',marker='<',linewidths=2)
plt.scatter(x,y2,color='r',label='Assualt',marker='o',linewidths=2)
plt.scatter(x,y3,color='m',label='UrbanPop',marker='*',linewidths=2)
plt.scatter(x,y4,color='c',label='Rape',marker='+',linewidths=2)

plt.xlabel('City')
plt.ylabel('Rape')
plt.title('Crime Data')

plt.legend(loc=1)

plt.show()
```
