```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
import scikit
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.PairGrid(iris) # this way it will only gives Grids

#for PairGrid we should always consider scatter plot for the map rather than any other
```

```python
x = sns.PairGrid(iris)

x.map(plt.scatter)

plt.show()
```

```python
x = sns.PairGrid(iris)

x.map(plt.bar)

plt.show()
```

```python
x = sns.PairGrid(iris)

x.map_diag(plt.hist)

plt.show()
```

```python
x = sns.PairGrid(iris,hue='species')

x.map_diag(plt.hist)
x.map_offdiag(plt.scatter)

x.add_legend()
plt.show()
```

```python
x = sns.PairGrid(iris,palette='winter_r',hue='species')

x.map_offdiag(plt.scatter)
x.map_diag(plt.hist,histtype='step',linewidth=4)

plt.show()
```
