```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
languages = ['Python','Java','R','Koltin','C','C++','Ruby','Perl','Scala','Go']
student1 = np.random.randint(9,90,(10))
student2 = np.random.randint(4,40,(10))
student3 = np.random.randint(6,60,(10))
langIndex=np.arange(len(languages))
width=0.2

print(student1)
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')

plt.bar(langIndex,student2,color='r',width=width,label='Student2')
plt.bar(langIndex+width,student1,color='b',width=width,label='Student1')
plt.bar(langIndex+width+width,student3,width,color='y',label='Student3')

plt.xticks(langIndex+width,languages,rotation=20,fontsize=13,color='k') # ticks method is use to replace the axis 
#plt.yticks(student1,langIndex)

plt.title("Students in Language classes")
plt.xlabel('Languages')
plt.ylabel('Number of Students')

plt.grid(color='k',linewidth=2,linestyle=':')

plt.legend(loc=1)
plt.show()
```
