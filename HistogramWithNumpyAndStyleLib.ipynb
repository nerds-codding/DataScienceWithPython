```python
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib import style

style.use('ggplot')
```

```python
age1 = np.random.randint(30,90,[100])
age2 = np.random.randint(40,100,[100])
```

```python
print(age1)
print(age2)
bins = [x for x in range(30,101,10)]
```

```python
plt.hist(age1,bins,color='m',rwidth=0.9,label='Age Frequency')# bins = range for the bars
# r width = gap between the bars 
# to use the label in the hist or any plot we have to use legend at the end

plt.grid(color='k',linewidth=1)

plt.xlabel('Age')
plt.ylabel('Frequency',fontsize=12,color='k')
plt.title('Histogram')
plt.legend(loc=1)
plt.show()
```

```python
plt.figure(figsize=(16,9)) # to override the graph size
plt.hist([age1,age2],bins,color=['m','g'],rwidth=0.9,label=['Age 1','Age 2'])
# bins = range for the bars
# r width = gap between the bars 
# to use the label in the hist or any plot we have to use legend at the end

plt.grid(color='k',linewidth=1,linestyle = ':')

plt.xlabel('Age')
plt.ylabel('Frequency',fontsize=12,color='k')
plt.title('Histogram')
plt.legend(loc=1)
plt.show()
```
